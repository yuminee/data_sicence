```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data/titanic.csv')
```

```python
# 타이타닉의 승객 분포도
```

```python
df.plot(kind = 'hist', y = 'Age', bins = 30)
```

```python
# 가장 높은 요금을 낸 사람의 나이대?
```

```python
df.plot(kind = 'scatter', x = 'Age', y = 'Fare')
```

```python
#생존자와 사망자
```

```python
df['Survived'].value_counts()
```

```python
#1등실, 2증실, 3등실 탑승객 수?
```

```python
df['Pclass'].value_counts()
```

```python
#생존율이 가장 높은 객실 등급?
```

```python
sns.kdeplot(df['Pclass'], df['Survived'])
```

```python
# Survived가 1이면 생존, 0이면 사망
# 확실히 3등실은 아래쪽에 밀집, 1등실은 위쪽에 밀집 되어있는것을 볼 수 있다.
# 겹쳐진 scatter의 정보를 알고싶을때에는 seabron의 kde를 사용한다.
```

```python
#나이별 생존율
```

```python
sns.stripplot(data = df, x = "Survived", y = "Age")
```

```python
sns.violinplot(data= df, x = 'Survived', y = 'Age')
```

```python
#나이와 생존율의 연관성은 없어보인다.
```

```python
#나이와 성별 생존율?
```

```python
sns.stripplot(data = df, x = "Survived", y = "Age", hue = 'Sex')
```
